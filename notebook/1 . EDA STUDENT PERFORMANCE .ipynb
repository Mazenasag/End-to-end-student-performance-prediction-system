### 3.8 Adding columns for "Total Score" and "Average"

```python
df['total score'] = df['math_score'] + df['reading_score'] + df['writing_score']
df['average'] = df['total score']/3
df.head()
```

```python
reading_full = df[df['reading_score'] == 100]['average'].count()
writing_full = df[df['writing_score'] == 100]['average'].count()
math_full = df[df['math_score'] == 100]['average'].count()

print(f'Number of students with full marks in Maths: {math_full}')
print(f'Number of students with full marks in Writing: {writing_full}')
print(f'Number of students with full marks in Reading: {reading_full}')
```

```python
reading_less_20 = df[df['reading_score'] <= 20]['average'].count()
writing_less_20 = df[df['writing_score'] <= 20]['average'].count()
math_less_20 = df[df['math_score'] <= 20]['average'].count()

print(f'Number of students with less than 20 marks in Maths: {math_less_20}')
print(f'Number of students with less than 20 marks in Writing: {writing_less_20}')
print(f'Number of students with less than 20 marks in Reading: {reading_less_20}')
```

#####  Insights
 - From above values we get students have performed the worst in Maths 
 - Best performance is in reading section

#### 4.3 Multivariate analysis using pieplot

```python
import matplotlib.pyplot as plt

# Set global figure size
plt.rcParams['figure.figsize'] = (20, 18)  # Adjusted for 3x2 layout

# Define subplot layout (3 rows, 2 columns)
fig, ax = plt.subplots(3, 2, figsize=(20, 18))  
ax = ax.flatten()  # Flatten to make looping easier

# Improved color palette for better visualization
colors = [
    ['#E63946', '#1D3557'],  # Gender (Red & Blue)
    ['#457B9D', '#E63946', '#F4A261', '#2A9D8F', '#264653'],  # Race/Ethnicity
    ['#F4A261', '#2A9D8F'],  # Lunch (Orange & Teal)
    ['#D7263D', '#3A86FF'],  # Test Course (Red & Blue)
    ['#A8DADC', '#457B9D', '#E63946', '#F4A261', '#2A9D8F', '#264653']  # Parental Education
]

# Titles for each pie chart
titles = ['Gender', 'Race/Ethnicity', 'Lunch', 'Test Course', 'Parental Education']

# Column names
columns = [
    'gender',
    'race_ethnicity',
    'lunch',
    'test_preparation_course',
    'parental_level_of_education'
]

# Generate Pie Charts
for i, col in enumerate(columns):
    size = df[col].value_counts()
    labels = size.index  # Dynamically fetch labels
    
    ax[i].pie(size, colors=colors[i], labels=labels, autopct='%1.1f%%',
    startangle=140, wedgeprops={'edgecolor': 'black'}, textprops={'fontsize': 14})
    ax[i].set_title(titles[i], fontsize=20, fontweight='bold')
    ax[i].axis('equal')  # Ensures a circular pie chart

# Hide empty subplot if dataset has fewer than 6 categories
for j in range(len(columns), len(ax)):
    fig.delaxes(ax[j])  # Remove extra subplots

# Improve layout
plt.tight_layout()

# Display the plot
plt.show()
```

```python
df['gender'].value_counts()
```

#####  Insights
- Number of Male and Female students is almost equal
- Number students are greatest in Group C
- Number of students who have standard lunch are greater
- Number of students who have not enrolled in any test preparation course is greater
- Number of students whose parental education is "Some College" is greater followed closely by "Associate's Degree"

#### BIVARIATE ANALYSIS ( Is gender has any impact on student's performance ? ) 

```python
gender_group = df.groupby('gender')[['math_score', 'reading_score']].mean()
gender_group
```

```python
gender_group['average'] = gender_group[['math_score', 'reading_score']].mean(axis=1)
gender_group
```

```python
plt.figure(figsize=(10, 8))

X = ['Total Average', 'Math Average']

# Get the scores for both male and female
female_scores = [gender_group['average']['female'], gender_group['math_score']['female']]
male_scores = [gender_group['average']['male'], gender_group['math_score']['male']]

# Define X axis positions
X_axis = np.arange(len(X))

# Plot the bar charts
plt.bar(X_axis - 0.2, male_scores, 0.4, label='Male')
plt.bar(X_axis + 0.2, female_scores, 0.4, label='Female')

# Set X axis labels and title
plt.xticks(X_axis, X)
plt.ylabel("Marks")
plt.title("Total Average vs Math Average Marks of Both Genders", fontweight='bold')

# Add legend and show the plot
plt.legend()
plt.show()
```

#### Insights 
- On an average females have a better overall score than men.
- whereas males have scored higher in Maths.
